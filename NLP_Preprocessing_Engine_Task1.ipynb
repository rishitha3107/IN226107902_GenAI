### 1. What is the difference between **“Love”** and **“love”** in NLP?

In Natural Language Processing (NLP), **“Love”** and **“love”** are treated as two different tokens because computers consider uppercase and lowercase letters separately. This increases the vocabulary size unnecessarily and may reduce model performance.

To avoid this issue, text is usually converted into lowercase during preprocessing so that both words are treated as the same token. This process is called **case normalization**.

---

### 2. What happens if stopwords are not removed?

If stopwords are not removed:

* The vocabulary size becomes larger
* Processing time increases
* Noise in the dataset increases
* Important keywords receive less importance during model training

However, removing stopwords depends on the NLP task and is not always recommended.

---

### 3. Provide two real-world scenarios where removing stopwords can be harmful

**Scenario 1: Sentiment Analysis**

Example sentence:

> "I am not happy"

If the stopword **“not”** is removed, the sentence becomes:

> "I happy"

This changes the meaning completely and produces incorrect sentiment results.

---

**Scenario 2: Question Answering Systems**

Example sentence:

> "What is the capital of India?"

Removing stopwords like **what**, **is**, **the**, and **of** makes the sentence unclear and harder for the system to interpret correctly.

### 4. Explain the difference between stemming and lemmatization


| Feature  | Stemming             | Lemmatization              |
| -------- | -------------------- | -------------------------- |
| Approach | Rule-based trimming  | Vocabulary + grammar-based |
| Output   | May not be real word | Always meaningful word     |
| Example  | running → runn       | running → run              |
| Speed    | Faster               | Slower                     |
| Accuracy | Lower                | Higher                     |




```python
# Task 2: Advanced Preprocessing Function

import re

def preprocess_text(text):

    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ("no", "not")]

    # Recreate cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


# Example test
example = "WOW!!! This is soooo coool!!! Visit http://example.com now"
tokens, cleaned = preprocess_text(example)

print("Tokens:", tokens)
print("Cleaned sentence:", cleaned)
```

```python
# Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("STRESS TEST RESULTS\n")

for i, sentence in enumerate(test_sentences, start=1):

    tokens, cleaned_sentence = preprocess_text(sentence)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```

```python
# Task 4: Token Analytics

def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0

    return total, unique, avg_len


print("\n========== TOKEN ANALYSIS ==========")

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)

    print("\nOriginal sentence:", sentence)
    print("Cleaned sentence:", clean)
    print("Total tokens:", total)
    print("Unique tokens:", unique)
    print("Avg token length:", round(avg_len, 2))
```

```python
# Task 5: Frequency Analysis

from collections import Counter

all_tokens = []

# Combine tokens from all processed sentences
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Count word frequency
word_frequency = Counter(all_tokens)

# Top 10 most frequent words
top_10_words = word_frequency.most_common(10)

# Top 5 least frequent words
least_5_words = word_frequency.most_common()[:-6:-1]

print("Top 10 Most Frequent Words:\n")
for word, freq in top_10_words:
    print(f"{word} : {freq}")

print("\nTop 5 Least Frequent Words:\n")
for word, freq in least_5_words:
    print(f"{word} : {freq}")
```

```python
# Task 6: Build Full Pipeline Function

def full_pipeline(text_list):
    """
    Processes a list of text inputs using preprocess_text()
    Returns:
    {
        "tokens": [...],
        "clean_sentences": [...]
    }
    """

    # Error handling: input must be a list
    if not isinstance(text_list, list):
        raise ValueError("Input must be a list of strings")

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, cleaned_sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example test run

pipeline_output = full_pipeline(test_sentences)

print("All Tokens:\n", pipeline_output["tokens"])
print("\nClean Sentences:\n", pipeline_output["clean_sentences"])
```

```python
# Task 7: Error Handling Tests

error_test_cases = [
    "",              # Empty string
    "😍🔥😂",         # Only emojis
    "1234567890"     # Only numbers
]

print("ERROR HANDLING TEST RESULTS\n")

for i, text in enumerate(error_test_cases, start=1):

    tokens, cleaned_sentence = preprocess_text(text)

    print(f"Test Case {i}")
    print("Original Text   :", repr(text))
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```

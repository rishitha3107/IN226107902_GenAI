```python
# imports
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

#### Loading Pretrained Model (FLAN-T5)

```python

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tok = AutoTokenizer.from_pretrained("google/flan-t5-base")
mod = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
```

#### Chat Function

```python
def chat():
    print("chatbot: hello! i am your ai assistant (type 'exit' to stop)")
    
    history = ""
    
    while True:
        
        usr = input("you: ")
        
        if usr.lower() in ['exit', 'quit']:
            print("chatbot: goodbye!")
            break
        
        if usr.strip() == "":
            print("chatbot: type a message")
            continue
        
        # FEW-SHOT PROMPT 
        prompt = f"""
You are a helpful and knowledgeable Artificial Intelligence (AI) assistant, you know all the correct factual answers and you can answer with intelligence and clarity, and you know how to greet properly as well

Follow these examples carefully:
User: What can you do?
Assistant: I can help answer questions, explain concepts, and assist with basic tasks or information.
User: Hello
Assistant: Hello! Nice to meet you. How can I assist you today?

User: What is Artificial Intelligence?
Assistant: AI is the ability of machines to perform tasks that normally require human intelligence, such as understanding, reasoning, and decision-making.
It enables systems to analyze information, learn from it, and act in a meaningful way.
User: Who created Python?
Assistant: Python was created by Guido van Rossum and released in 1991.

User: Thank you
Assistant: You're welcome! Feel free to ask more questions.

Now continue the conversation:

{history}
User: {usr}
Assistant:
"""
        
        inp = tok(prompt, return_tensors="pt")
        
        out = mod.generate(
            **inp,
            max_new_tokens=50,
            do_sample=False   
        )
        
        res = tok.decode(out[0], skip_special_tokens=True)
        res = res.split("Assistant:")[-1].strip().split("\n")[0]
        
        print("chatbot:", res)
        
        history += f"User: {usr}\nAssistant: {res}\n"
```

```python
chat()
```
